4) WAP in python has a class Circle. Use a class variable that defines the value
of constant Pie value=3.14. Use this class variable to calculate the
area and circumference of the circle with specified radius.

```python
class Circle():
    def __init__(self, r):
        self.radius = r

    def area(self):
        return self.radius**2*3.14
    
    def perimeter(self):
        return 2*self.radius*3.14


NewCircle = Circle(15)

print(NewCircle.area())

print(NewCircle.perimeter())
```

Triangle from polygon and write methods to get the details of their dimensions and
hence calculate the area.

```python
class Polygon:
   def __init__(self, n):
        self.number_of_sides = n
  
   def print_num_sides(self):
        print('There are ' + str(self.number_of_sides) + ' sides.')

class Rectangle(Polygon):
   def __init__(self, lenghts_of_sides):
        Polygon.__init__(self, 4)
        self.lengths_of_sides = lenghts_of_sides  

   def get_area(self):
        x, y = self.lenghts_of_sides
        return x * y

 
class Triangle(Polygon):
   def __init__(self, lengths_of_sides):
        Polygon.__init__(self, 3)
        self.lengths_of_sides = lengths_of_sides 

   def get_area(self):
        a, b, c = self.lengths_of_sides
        s = (a + b + c) / 2
        return (s*(s-a)*(s-b)*(s-c)) ** 0.5

p = Polygon(8)
p.print_num_sides()

tri = Triangle([6, 7, 8])
print("area of triangle :",tri.get_area())

tri.print_num_sides()
```
